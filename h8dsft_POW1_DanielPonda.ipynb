```python
# MENAMPILKAN DATASET DARI FILE EXCEL MENGGUNAKANA PANDAS
import numpy as np
import pandas as pd
x = pd.read_csv("Pokemon.csv")
x
```

```python
# DATA LOADING
x.head(5)
```

```python
# DATA LOADING
x.info()
```

```python
# MELIHAT DESKRIPSI STATISTIK DARI DATA SET YANG KITA MILIKI. SEPERTI JUMLAH DATA, MEAN, NILAI MAX, NILAI MIN, DLL
x.describe()
```

```python
# MELIHAT JUMLAH MISSING VALUE PADA SETIAP COLUMN
x.isnull().sum()
```

```python
# MENGISI MISSING VALUE DENGAN "None"
x["Type 2"].fillna("None", inplace = True)
x
```

```python
# MENGGANTI COLUMN "#" DENGAN NAMA "ID"
x.columns = x.columns.str.replace("#","ID")
x
```

```python
# DATA QUERY: MENAMPILKAN HP LEBIH KECIL DARI 100
x[ (x["HP"]> 100)]
```

```python
# DATA QUERY: MENAMPILKAN GENERATION DIATAS 3
x[ (x["Generation"]> 3)]
```

```python
# DATA QUERY: MENAMPILKAN SPEED YANG DIATAS 50
x[ (x["Speed"] < 50)]
```

```python
# MENGELOMPOKAN POKEMON YANG TIDAK LEGENDARY
x [ 
    (x["Legendary"] == False)
].groupby(["Generation"])["Legendary"].count()
```

```python
# MENGELOMPOKAN POKEMON YANG HANYA MEMILIKI 1 TIPE
x [ 
    (x["Type 2"] == "None")
].groupby(["Name", "Generation"])["Generation"].count()
```

```python
# MENGELOMPOKAN POKEMON FIRE
x [ 
    (x["Type 1"] == "Fire")&
    (x["Type 2"] == "None")
].groupby(["Generation","Legendary"])["HP"].count()
```

```python
x [ 
    (x["Legendary"] == False)
].groupby(["Generation"])["Legendary"].count().plot()
```

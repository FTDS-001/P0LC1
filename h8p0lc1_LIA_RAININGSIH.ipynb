```python
import pandas as pd

df = pd.read_excel('https://github.com/fahmimnalfrzki/Dataset/blob/main/NMC1-9.xlsx?raw=true')
print(df)
```

Menghitung resistansi untuk nilai cycle ganjil dan menghitung daya untuk nilai cycle genap pada dataframe.

```python
for index, row in df.iterrows(): # looping untuk setiap baris dalam df, function
  cycle = row['Cycle']           #isi variable cycle
  vm = row['V_m']                #isi variable vm
  current = row['Current (A)']   #isi variable current
  time = row['time(s)']          #isi variable time
  if cycle % 2 == 0:             #conditional if, jika genap
    P = vm * current / time      #hitung nilai daya baterai tiap waktu
    print(P)                     #print nilai daya baterai tiap waktu
  else:                          #jika ganjil
    R = vm/current               #hitung nilai resistansi
    print(R)                     #print nilai resistansi
```

```python
# MENAMPILKAN DATA MENGGUNAKAN PANDAS
import numpy as np
import pandas as pd
x = pd.read_excel("NMC1-9.xlsx")
x
```

```python
# MENGHITUNG HAMBATAN DAN DAYA
def hambatan():
    r= x["V_m"]/(x["Current (A)"])
    return r
def daya():
    p= x["V_m"]*(x["Current (A)"])
    return p
```

```python
# MENGHITUNG HAMBATAN JIKA CYCLE GANJIL, DAN MENGHITUNG DAYA JIKA CYCLE GENAP
for i in x["Cycle"]:
    if i%2  == 0:
        print(daya())
    else:
        print(hambatan())
```
